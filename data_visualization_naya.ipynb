```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# untuk membaca data csv
df = pd.read_csv("https://raw.githubusercontent.com/sushantag9/Supermarket-Sales-Data-Analysis/master/supermarket_sales%20-%20Sheet1.csv")
df.head()
```

```python
df.info()
```

```python
df["DateTime"] = pd.to_datetime(df["Date"] + " " + df["Time"])
df.head()
```

```python
df.info()
```

```python
df.drop(columns=["Date","Time"],inplace=True)
df.info()
```

# LINE PLOT

```python
# menyatakan dalam bentuk jam
df["hour"] = df[""].dt.hour
# menyatakan dalam bentuk minggu
df["weekday"] = df[""].dt.weekday
# menyatakan dalam bentuk bulan
df["month"] = df[""].dt.month
```

```python
df["weekday"] = df["DateTime"].dt.weekday
df_plot = df.groupby("weekday").agg({"Quantity":"sum"}).reset_index()
```

```python
df
```

```python
df_plot
```

```python
sns.lineplot(data=df_plot,x="weekday",y="Quantity")
plt.title("Quantity in Weekday")
plt.show()
```
